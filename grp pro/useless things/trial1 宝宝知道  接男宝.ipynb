```python
df1=baobaozhidao[(baobaozhidao['rs-line 2']=='接')&(baobaozhidao['rs-line 3']=='男宝')]
df1
```

```python
df1['rs-line 4']= df1['rs-line 4'].fillna('')
df1
```

```python
#drop the first two columns
to_drop=['rs-line 2', 'rs-line 3']
df1.drop(columns=to_drop,inplace=True)
df1.columns
```

```python
#change the chinese dates into numerical ones

def to_date(series):
    
    res = pd.to_datetime(series, errors='coerce')
    if pd.isna(res).sum() == 0:
        return res
    else:
        parts = series.astype(str).str.split('年|月|日', expand=True)
        # xxxx年xx月xx日
        dt = pd.to_datetime(parts[0]+'/'+parts[1]+'/'+parts[2], format='%Y/%m/%d', errors='coerce')
        res = res.fillna(dt)
        if pd.isna(res).sum() == 0:
            return res

df1['realtime'] = to_date(df1['time'])

df1
```

```python
#list the data frame with the timeorder
timeorder =df1.sort_values(by='realtime', ascending= True)
timeorder
```

```python
#adjust the column places

timeorder_realtime = timeorder.realtime
timeorder=timeorder.drop('time',axis=1)
timeorder = timeorder.drop('realtime',axis=1)
timeorder.insert(0,'realtime',timeorder_realtime)

timeorder
```
