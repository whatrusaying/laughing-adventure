```python
import pandas as pd
import numpy as np
from pandas import *
analysis=pd.read_csv(r'D:\1\international.csv')
analysis
```

# only select first 200 columns for analysis as the example 

```python
ana=analysis.head(200)
ana
```

```python
#mark and emo relationship
markemo=ana[['mark', 'emo']]
table1= markemo.sort_values(by='mark',ascending= True)
import matplotlib.pyplot as plt
table1.plot(kind='scatter',
            x='mark',
            y='emo')
```

```python
#emo and semester
sememo=ana[['semester', 'emo']]
sememo.groupby('semester').agg('mean').plot(kind='bar')
```

```python
#year and emo
ana[['year', 'emo']].groupby('year').agg('mean').plot(kind='line')
```

```python
#workload_num and emo
ana[['workload_num', 'emo']].plot(x='workload_num',y='emo',kind='scatter')
```
