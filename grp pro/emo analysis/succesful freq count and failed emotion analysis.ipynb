```python
import pandas as pd
import numpy as np
from textblob import TextBlob

df1= pd.read_table(r"D:\1\filepath.txt")

df1['body']=df1['body'].astype(str)
```

```python
from textblob import Word

df1['body'] = df1['body'].apply(
    lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df1['body'].head()
```

```python
from textblob.classifiers import NaiveBayesClassifier

train = [
    ('I love this.', 'pos'),
    ('This is an amazing course!', 'pos'),
    ('I feel very good about these assignments.', 'pos'),
    ('This is my best work.', 'pos'),
    ("What an awesome view", 'pos'),
    ('I do not like this course', 'neg'),
    ('I am tired of this stuff.', 'neg'),
    ("I can't deal with this", 'neg'),
    ('My boss is horrible.', 'neg'),
    ('horrible tutor',"neg"),
    ('still consider favourate course year','pos'),
    ('teacher nice, friendly supportive.','pos'),
    ('totally rubbish important study hard quizzes','neg'),
    ('as professor, gives grading without explanation refuses revaluate final papers.','neg')
    
    
]
test = [
    ('The beer was good.', 'pos'),
    ('I do not enjoy my job', 'neg'),
    ("horrible .", 'neg'),
    ("I feel amazing!", 'pos'),
    ('Gary is a teacher of mine.', 'pos')
]
```

```python
cl = NaiveBayesClassifier(train)
```

```python
def blob(Series):
    for items in Series:
        blob = TextBlob(items, classifier=cl)
        blob2 = blob.classify()
    return blob2

df1['body3']=df1['body'].apply(blob)
```
