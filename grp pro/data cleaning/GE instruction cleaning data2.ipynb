```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from pandas import Series
csv_file=pd.read_csv(r"D:\1\comments.csv",encoding='utf-8')
csv_file
```

```python
csv_file['chinese course name']= csv_file['chinese course name'].fillna("")
```

```python
#change the grade into numerical data
mark={'A+':4.30,
     'A':4.00,
     "A-":3.70,
     "B+":3.30,
     "B":3.00,
     "B-":2.70,
     "C+":2.30,
     "C":2.00,
     "C-":1.70,
     "D":1.00,
     "F":0.00}

csv_file['mark'] = csv_file['grade'].map(str).map(mark)
csv_file=csv_file.drop(['grade'],axis=1)
```

```python
workload_num={'非常輕鬆':0,
              '輕鬆':1,
              '一般':2,
              '繁重':3,
              '非常繁重':4}
#the heavier the workload is, the higher score it gets.
csv_file['workload_num'] =csv_file['workload'].map(str).map(workload_num)
csv_file = csv_file.drop(['workload'],axis=1)
```

```python
#separate semester and year
def convert_words(series):
    new_value = series.replace('Sem A','SemA')
    #return new_value  #if u cannot run the next block, add or delete this line,then you run this block again
    new_value = series.replace('Sem B','SemB')
    return new_value

csv_file['sem'] = csv_file['sem'].apply(convert_words)

csv_file.head()
```

```python
csv_file[['year','semester']] = csv_file['sem'].str.split(expand=True)

csv_file.drop('sem', axis=1, inplace=True)
```

```python
from datetime import datetime
csv_file['comment time'] = pd.to_datetime(csv_file['comment time'], format = "%b %d, %Y %H:%M", errors = 'coerce')
```
