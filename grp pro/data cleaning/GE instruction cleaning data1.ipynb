```python
import pandas as pd
import numpy as np
```

```python
comments=pd.read_csv(r'D:\1\ge.csv')
comments.head(5)
```

```python
comments.info()
```

```python
comments.columns
```

```python
#drop useless columns
to_drop=['btn href', 'dropdown-menu href', 'dropdown-menu href 2',
       'dropdown-menu href 3', 'dropdown-menu href 4', 'dropdown-menu href 5',
       'panel-comment-heading-am href']
comments.drop(columns=to_drop,inplace=True)
comments
```

```python
#change column names
comments=comments.rename(columns = {'panel-course-code':'course code'})
comments=comments.rename(columns = {'panel-course-title':'course name'})
comments=comments.rename(columns = {'panel-comment-grade':'grade'})
comments=comments.rename(columns = {'panel-comment-aside-meta 2':'sem'})
comments=comments.rename(columns = {'panel-comment-aside-meta 4':'teacher'})
comments=comments.rename(columns = {'panel-comment-aside-meta 6':'workload'})
comments=comments.rename(columns = {'text-muted':'comment time'})
comments=comments.rename(columns = {'panel-course-title 2':'chinese course name'})
comments.head(3)
```

```python
#clear empty data
comments['chinese course name']=comments['chinese course name'].fillna('')
comments['panel-body 2']=comments['panel-body 2'].fillna('')
comments.head(3)
```

```python
#gather the comments
comments['body']=comments['panel-body']+comments['panel-body 2']
comments=comments.drop(['panel-body'],axis=1)
comments=comments.drop(['panel-body 2'],axis=1)
comments.head()
```

```python
comments.to_csv('comments.csv')#保存到本地路径
```
